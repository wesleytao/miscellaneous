```python
# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cityofnewyork.us", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.cityofnewyork.us,
#                  MyAppToken,
#                  userame="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("6b8a-2fci",limit=4500)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
monthdate=set(results_df.yearmonth) 
monthdate=[date for date in monthdate if date not in ['FY 2010', 'FY 2011', 'FY 2012', 'FY 2013', 'FY 2014', 'FY 2015', 'FY 2016', 'FY 2017']]
```

```python
results_df.head()
```

```python
mysubset=results_df[results_df.incidentborough=="Citywide"][results_df.yearmonth.isin(monthdate)]
```

```python
mysubset=mysubset[["incidentclassification","incidentcount","yearmonth"]]
```

```python
totalcount_df=mysubset.pivot(index="yearmonth",columns="incidentclassification",values="incidentcount")
```

```python
totalcount_df.head()
```

```python
int("3")
```

```python
sumofallcount=pd.to_numeric(totalcount_df["All Fire/Emergency Incidents"])
False_Alarm=pd.to_numeric(totalcount_df["False Alarm"])/sumofallcount
Medical_False_Alarm=pd.to_numeric(totalcount_df["Medical False Alarm"])/sumofallcount +False_Alarm
NonStructural_Fires=pd.to_numeric(totalcount_df["NonStructural Fires"])/sumofallcount+Medical_False_Alarm
Medical_Emergencies=pd.to_numeric(totalcount_df["Medical Emergencies"])/sumofallcount+NonStructural_Fires
Structural_Fires=pd.to_numeric(totalcount_df["Structural Fires"])/sumofallcount+Medical_Emergencies
NonMedical_Emergencies=pd.to_numeric(totalcount_df["NonMedical Emergencies"])/sumofallcount+Structural_Fires
```

```python
new=pd.concat([False_Alarm, Medical_False_Alarm,NonStructural_Fires,Medical_Emergencies,Structural_Fires,NonMedical_Emergencies], axis=1)
new.index=[date.replace("/",".") for date in new.index]
```

```python
new.as_matrix()
```

```python
new=new.rename(index=str, columns={0:"FalseAlarm", 1: "MedicalFalseAlarm",2:"NonStructuralFires",3:"MedicalEmergencies",4:"StructuralFires",5:"NonMedicalEmergencies"})
```

```python
path=os.path.expanduser("~\python_project\Data Visualization\datapecent.csv")
new.to_csv(path)
```
