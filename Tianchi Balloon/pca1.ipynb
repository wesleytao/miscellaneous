```python
from sklearn.decomposition import PCA
```

```python
df = pd.DataFrame(data = samplemodel2)
```

```python
df1 = df.iloc[:,0:10]
df1.head()
```

```python
df1_centered = df1 - df1.mean(axis = 0)
```

```python
pca = PCA()
```

```python
pca.fit(df1_centered)
```

```python
cumsum = np.cumsum(pca.explained_variance_ratio_)
```

```python
d = np.argmax(cumsum >= 0.95) + 1
d
```

```python
var = pca.explained_variance_ratio_
var
```

```python
plt.plot(var)
```

```python
pca = PCA(n_components=1)

pca.fit(df1_centered)

x=pca.fit_transform(scale(df1_centered))

print (x)
```
